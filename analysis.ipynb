```python
import pandas as pd
import numpy as np 
import os
```

```python
data_first= pd.read_csv('/home/rajesh/work/ansc_ta/codes/files/ANSC-108-502 (1).csv')
```

```python
main_emails= data_first['Email'].unique()
main_emails= [email.split('@')[0] for email in main_emails]
```

```python
import sys
from thefuzz import fuzz

def find_unmatched_emails(main_list, comparison_list, threshold=60):
    """
    Finds items in main_list that do not have a match in
    comparison_list with a similarity score >= threshold.
    """
    non_matching_emails = []

    # Loop through every email in your main list
    for m_email in main_list:
        max_similarity =0
        
        # Compare it against every email in the second list
        for c_email in comparison_list:
            similarity = fuzz.ratio(m_email, c_email)
            
            # Keep track of the highest match found
            if similarity > max_similarity:
                max_similarity = similarity
        
        # If the BEST match found is STILL less than the threshold,
        # then this email is a "non-match"
        if max_similarity < threshold:
            non_matching_emails.append(m_email)
            
    return non_matching_emails
```

```python
for files in os.listdir('/home/rajesh/work/ansc_ta/codes/files'):
    path=  os.path.join('/home/rajesh/work/ansc_ta/codes/files', files)
    file= pd.read_csv(path)
    try:
        emails=file['Your University Email Address'].unique()
    except: "key error"
    emails=[email.split('@')[0] for email in emails]
    non_matching_email=find_unmatched_emails(main_emails, emails)
    non_matching_email= [str(email)+str('@email.tamu.edu') for email in non_matching_email]
    names=data_first[data_first['Email'].isin(non_matching_email)]['Name'].unique()
    print(f"file {path} and unmatched email {names} ")
```
